```python
import skmob
from skmob.preprocessing import detection, clustering, compression
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../Trips.csv')
df.head()
```

```python
df['Time'] = (pd.to_datetime(df['time'], format='%m/%d/%Y %H:%M:%S').dt.time)
```

```python
df['9am'] = np.where(((df['Time'] < pd.to_datetime('9:00:00').time()) | (df['Time'] > pd.to_datetime('12:00:00').time())) , 'No', 'Yes')
```

```python
df = df[df['9am'] == 'Yes']
df.to_csv('Nineam.csv')
df.head()
```
